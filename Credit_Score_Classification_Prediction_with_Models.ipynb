```python
# Mapping the target columns for preparing the model.
data["credit_mix"]= data["credit_mix"].map({"Standard":1, "Good":2, "Bad":0})
data["credit_score"]= data["credit_score"].map({"Standard":1, "Good":2, "Poor":0})
```

```python
# Droping null row.
null_rows = data[data.isnull().any(axis=1)]
null_rows

index_to_drop = null_rows.index[0]
data = data.drop(index_to_drop)
```

```python
data["credit_score"]= data["credit_score"].astype(int)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

x= np.array(data[["annual_income", "monthly_inhand_salary",
                   "num_bank_accounts", "num_credit_card",
                   "interest_rate", "num_of_loan",
                   "delay_from_due_date", "num_of_delayed_payment",
                   "credit_mix", "outstanding_debt",
                   "credit_history_age", "monthly_balance"]])
y= np.array(data[["credit_score"]])
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, train_size=0.70, random_state=42)
```

```python
## Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
model.fit(xtrain, ytrain)
```

```python
y_predit= model.predict(xtest)
print(y_predit)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(ytest, y_predit))
print(classification_report(ytest, y_predit))
```
